```python
#Descarga de datos (grupos y equipos de cada grupo antes de empezar el torneo): 

todas_tablas = pd.read_html('https://web.archive.org/web/20221115040351/https://en.wikipedia.org/wiki/2022_FIFA_World_Cup')
```

```python
#Creando df y diccionario:

dict_tablas = {}

for letra, i in zip(alfabeto, range(12,62, 7)):
    df = todas_tablas[i]
    df.rename(columns={df.columns[1] : 'Team'}, inplace = True)
    df.pop('Qualification')
    dict_tablas[f'Grupo {letra}']= df
```

```python
#Data set todos los años:
from bs4 import BeautifulSoup
import requests

years = [1930, 1934, 1938, 1950, 1954, 1958, 1962, 1966, 1970, 1974, 1978, 1982, 1986, 1990, 1994, 1998, 2002, 2006, 2010, 2014, 2018]

#Web scraping usando BeatufulSoup

#web = 'https://en.wikipedia.org/wiki/2014_FIFA_World_Cup'    #Dirección de la web de donde vamos a extrar los datos
 
#response = requests.get(web)  #Realizando la petición
#content = response.text       #Guardando la información de la petición en texto

#soup = BeautifulSoup(content, 'lxml') #Permite la mejor lectura de los datos


#Data de todos los mundiales (para cualquier año):

def get_matches(year):

    web = f'https://en.wikipedia.org/wiki/{year}_FIFA_World_Cup'    #Dirección de la web de donde vamos a extrar los datos
    
    response = requests.get(web)  #Realizando la petición
    content = response.text       #Guardando la información de la petición en texto

    soup = BeautifulSoup(content, 'lxml') #Permite la mejor lectura de los datos
    
    #Extraccion de datos:

    matches = soup.find_all('div', class_='footballbox')

    home = []
    score = []
    away = []

    for match in matches:
        home.append(match.find('th', class_='fhome').get_text())
        score.append(match.find('th', class_='fscore').get_text())
        away.append(match.find('th', class_='faway').get_text())

    dict_football = {'home': home,'score': score ,'away': away}

    df_football = pd.DataFrame(dict_football)

    df_football['year'] = year

    return df_football
```

```python
#Extraer data de todos los mundiales:

fifa = [get_matches(year) for year in years]

df_fifa = pd.concat(fifa, ignore_index=True)

#df_fifa.to_csv('fifa_worldcup_hitorical_data.csv', index = False) #Archivo final
```

```python
#Data para el mundial del 2022, previo a que se jugara:

web_2022 = 'https://web.archive.org/web/20221115040351/https://en.wikipedia.org/wiki/2022_FIFA_World_Cup'

response_2022 = requests.get(web_2022)  #Realizando la petición
content_2022 = response_2022.text       #Guardando la información de la petición en texto

soup_2022 = BeautifulSoup(content_2022, 'lxml') #Permite la mejor lectura de los datos



#Extraccion de datos:

matches_2022 = soup_2022.find_all('div', class_='footballbox')

home_2022 = []    
score_2022 = []
away_2022= []

for match in matches_2022:
    home_2022.append(match.find('th', class_='fhome').get_text())
    score_2022.append(match.find('th', class_='fscore').get_text())
    away_2022.append(match.find('th', class_='faway').get_text())

dict_football_2022 = {'home': home_2022,'score': score_2022 ,'away': away_2022}

df_football_2022= pd.DataFrame(dict_football_2022)

df_football_2022['year'] = '2022'

df_football_2022.to_csv('fifa_worldcup_fixture.csv', index = False)
```
